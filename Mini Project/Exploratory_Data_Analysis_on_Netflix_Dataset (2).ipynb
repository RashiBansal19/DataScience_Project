# 4. Data Cleaning

### 1. Missing Values

```python
netflix.isnull().sum()
```

### 2. Handle Missing Values

```python
netflix['director'].fillna('Unknown', inplace=True)
netflix['cast'].fillna('Not Available', inplace=True)
netflix['country'].fillna(netflix['country'].mode()[0], inplace=True)
netflix['rating'].fillna(netflix['rating'].mode()[0], inplace=True)

netflix.dropna(subset=['date_added'], inplace=True)
```

### 3. Remove Duplicates

```python
netflix.duplicated().sum()

netflix.drop_duplicates(inplace=True)
```

### 4. Convert Data Types

```python
# Convert date_added to datetime
netflix['date_added'] = netflix['date_added'].str.strip()
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce')

# Extract year and month
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added'] = netflix['date_added'].dt.month
```

# 5. Outlier Detection

### 1. Boxplot

```python
sns.boxplot(x=netflix['release_year'])
plt.title("Outlier Detection - Release Year")
plt.show()
```

### 2. IQR Method

```python
Q1 = netflix['release_year'].quantile(0.25)
Q3 = netflix['release_year'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

netflix = netflix[(netflix['release_year'] >= lower) &
        (netflix['release_year'] <= upper)]
```

# 6. Univariate Analysis

### 1. Content Type Distribution

```python
sns.countplot(x='type', data=netflix)
plt.title("Movies vs TV Shows")
plt.show()
```

### 2. Rating Distribution

```python
plt.figure(figsize=(12,6))
sns.countplot(y='rating', data=netflix,
              order=netflix['rating'].value_counts().index)
plt.title("Content Ratings")
plt.show()
```

### 3. Release Year Distribution

```python
sns.histplot(netflix['release_year'], bins=30, kde=True)
plt.title("Release Year Distribution")
plt.show()
```

### 4. Top 10 Countries

```python
netflix['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Countries")
plt.show()
```

# 7. Bivariate Analysis

### 1. Type vs Rating

```python
sns.countplot(x='rating', hue='type', data=netflix)

plt.xticks(rotation=45)
plt.title("Rating by Content Type")
plt.show()
```

### 2. Content Added Over Years

```python
sns.countplot(x='year_added', data=netflix)

plt.xticks(rotation=90)
plt.title("Content Added Each Year")
plt.show()
```

### 3. Country vs Type

```python
top_countries = netflix['country'].value_counts().head(10).index

temp = netflix[netflix['country'].isin(top_countries)]

sns.countplot(y='country',
              hue='type',
              data=temp)

plt.title("Country vs Content Type")
plt.show()
```

# 8. Correlation Analysis

```python
numeric_df = netflix.select_dtypes(include=np.number)

corr = numeric_df.corr()

plt.figure(figsize=(8,5))

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

#### Observation

* Positive correlation between year_added and release_year.
* No strong negative correlation found.

# 9. Feature Engineering

### 1. Content Age

```python
netflix['content_age'] = 2024 - netflix['release_year']
```

### 2. Content Category

```python
netflix['content_category'] = np.where(
    netflix['content_age'] <= 5,
    'Recent',
    'Old'
)

netflix[['release_year',
    'content_age',
    'content_category']].head()
```
