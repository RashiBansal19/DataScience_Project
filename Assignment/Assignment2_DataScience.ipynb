Q22. Apply multiple conditions using & and |.

```python
d16 = {
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"],
    "Score": [85, 92, 78, 88]
}

df16 = pd.DataFrame(d16)

print(df16[(df16["City"] == "Delhi") & (df16["Score"] > 80)])

print(df16[(df16["City"] == "Delhi") | (df16["Score"] > 90)])
```

Q36. Merge two DataFrames using a common key.

```python
a1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Aman", "Riya", "Karan"]
})

a2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Score": [85, 92, 78]
})

print(pd.merge(a1, a2, on="ID"))
```

Q37. Perform a left join between two DataFrames.

```python
b1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Aman", "Riya", "Karan"]
})

b2 = pd.DataFrame({
    "ID": [1, 2],
    "Score": [85, 92]
})

print(pd.merge(b1, b2, on="ID", how="left"))
```

Q40. Mini Project: Create a sales DataFrame and answer business questions using filtering, groupby, and aggregation.

```python
sales = {
    "Product": ["Pen", "Book", "Pen", "Bag", "Book"],
    "City": ["Delhi", "Delhi", "Mumbai", "Delhi", "Mumbai"],
    "Quantity": [10, 5, 15, 2, 8],
    "Price": [10, 50, 10, 500, 50]
}

df30 = pd.DataFrame(sales)

df30["Total"] = df30["Quantity"] * df30["Price"]

print("Sales Data:")
print(df30)

print("\nProducts sold in Delhi:")
print(df30[df30["City"] == "Delhi"])

print("\nTotal Sales by City:")
print(df30.groupby("City")["Total"].sum())

print("\nAverage Sales by Product:")
print(df30.groupby("Product")["Total"].mean())

print("\nOverall Total Sales:")
print(df30["Total"].sum())
```
